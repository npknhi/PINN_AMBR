# AMBR Real Data Plots

Plots real measurements from `ambr_preprocessed.csv` for every `Experiment_id`.

```python
import sys

import matplotlib.pyplot as plt

sys.path.insert(0, "../..")
from src.utils.preprocess import load_preprocessed_ambr_data, get_experiment_ids, plot_ambr_experiment
```

```python
frame = load_preprocessed_ambr_data("ambr_preprocessed.csv")
ids = get_experiment_ids(frame)

print(f"Rows: {len(frame):,}")
print(f"Experiments: {len(ids)}")
print(ids)
```

```python
for experiment_id in ids:
    fig = plot_ambr_experiment(frame, experiment_id)
    plt.show()
    plt.close(fig)
```
